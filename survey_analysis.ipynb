```python
import pandas as pd
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize':(20,10)})
```

```python
df = pd.read_csv("data/survey.csv").drop("Sygnatura czasowa", axis=1)
photos = ["zdjęcie 1", "zdjęcie 2", "zdjęcie 3", "zdjęcie 4", "zdjęcie 5"]
methods = ["seam carving", "vision-stretcher", "GAN outpainting", "OutpainterNet"]
mi = pd.MultiIndex.from_product([photos, methods], names=["zdjęcie", "metoda"])
df.columns = mi
```

```python
df
```

```python
ax = sns.barplot(data=pd.melt(df, value_name="oceny"), x="zdjęcie", y="oceny", hue="metoda")
for container in ax.containers:
    ax.bar_label(container, fmt="%.2f", label_type="center")
```

```python
ax = sns.barplot(data=pd.melt(df, value_name="oceny"), x="zdjęcie", y="oceny", hue="metoda", errorbar="se")
for container in ax.containers:
    ax.bar_label(container, fmt="%.2f", label_type="center")
```

```python
ax = sns.barplot(data=pd.melt(df, value_name="oceny"), x="zdjęcie", y="oceny", hue="metoda", errorbar="ci")
for container in ax.containers:
    ax.bar_label(container, fmt="%.2f", label_type="center")
```

```python
ax = sns.barplot(data=pd.melt(df, value_name="oceny"), x="metoda", y="oceny")
for container in ax.containers:
    ax.bar_label(container, fmt="%.2f", label_type="center")
```
